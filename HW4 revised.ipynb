# HW4: Linear Regression

```python
import numpy as np
from scipy import stats
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#grab data 
df=pd.read_excel('LAmeantempdata.xlsx')
data=df.to_numpy()
#set up for look to parse through months Jan-Dec
for i in range (1,13):
    if i==1:
        print("January")
    elif i==2:
        print("\n February")
    elif i==3:
        print("\n March")
    elif i==4:
        print("\n April")
    elif i==5:
        print("\n May")
    elif i==6:
        print("\n June")
    elif i==7:
        print("\n July")
    elif i==8:
        print("\n August")
    elif i==9:
        print("\n September")
    elif i==10:
        print("\n October")
    elif i==11:
        print("\n November")
    elif i==12:
        print("\n December")
    else:
        print("error")
    x=data[:,0]
    y=data[:,i]
    #set up regression variables for FULL DATA
    ind=(~np.isnan(y)) & (x > 1878) & (x < 2020)
    D=[x[ind],y[ind]]
    c=np.cov(D)
    a=c[0,1]/c[0,0]
    b=np.mean(y[ind])-a*np.mean(x[ind])
    #plot linear regression, coefficient of determination r2a
    plt.plot(x,y)
    plt.plot(x[ind],a*x[ind]+b,'r--')
    plt.show()
    r2a=c[0,1]**2/c[0,0]/c[1,1]
    print([a,b,r2a])
    #set up regression variables for 1990-2020
    ind=(~np.isnan(y)) & (x > 1990) & (x < 2020)
    D=[x[ind],y[ind]]
    c=np.cov(D)
    a=c[0,1]/c[0,0]
    b=np.mean(y[ind])-a*np.mean(x[ind])
    #plot linear regression, coefficient of determination r2b
    plt.plot(x,y)
    plt.plot(x[ind],a*x[ind]+b,'r--')
    plt.show()
    r2b=c[0,1]**2/c[0,0]/c[1,1]
    print([a,b,r2b])
    print("R2 for full data is equal to R2 for 1990-2020: ", r2a==r2b)
```
