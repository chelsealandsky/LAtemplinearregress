# Homework 4 - Linear Regression

```python
import numpy as np
from scipy import stats
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#grab data for jan
df=pd.read_excel('LAmeantempdata.xlsx')
data=df.to_numpy()
for i in range (0,13):
    x=data[:,0]
    y=data[:,i]
    ind=(~np.isnan(y)) & (x > 1878) & (x < 2020)
    D=[x[ind],y[ind]]
    c=np.cov(D)
    a=c[0,1]/c[0,0]
    b=np.mean(y[ind])-a*np.mean(x[ind])
    plt.plot(x,y)
    plt.plot(x[ind],a*x[ind]+b,'r--')
    plt.show()
    r2a=c[0,1]**2/c[0,0]/c[1,1]
    print([a,b,r2a])
    ind=(~np.isnan(y)) & (x > 1990) & (x < 2020)
    D=[x[ind],y[ind]]
    c=np.cov(D)
    a=c[0,1]/c[0,0]
    b=np.mean(y[ind])-a*np.mean(x[ind])
    plt.plot(x,y)
    plt.plot(x[ind],a*x[ind]+b,'r--')
    plt.show()
    r2b=c[0,1]**2/c[0,0]/c[1,1]
    print([a,b,r2b])
```

```python
#set up regression variables for FULL DATA
ind=(~np.isnan(y)) & (x > 1878) & (x < 2020)
D=[x[ind],y[ind]]
c=np.cov(D)
a=c[0,1]/c[0,0]
b=np.mean(y[ind])-a*np.mean(x[ind])
```

```python
#plot linear regression, coefficient of determination r2a
plt.plot(x,y)
plt.plot(x[ind],a*x[ind]+b,'r--')
plt.show()
r2a=c[0,1]**2/c[0,0]/c[1,1]
print([a,b,r2a])
```

```python
#set up regression variables for 1990-2020
ind=(~np.isnan(y)) & (x > 1990) & (x < 2020)
D=[x[ind],y[ind]]
c=np.cov(D)
a=c[0,1]/c[0,0]
b=np.mean(y[ind])-a*np.mean(x[ind])
```

```python
#plot linear regression, coefficient of determination r2b
plt.plot(x,y)
plt.plot(x[ind],a*x[ind]+b,'r--')
plt.show()
r2b=c[0,1]**2/c[0,0]/c[1,1]
print([a,b,r2b])
```

```python
r2a==r2b
```
